```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf
```

```python
data = pd.read_excel("Ticker_data_sp500.xlsx", sheet_name= None)
print(data.keys())
ticker_data = data["top_50"]
print(ticker_data.columns)
sp_500 = "^GSPC"
tickers = ticker_data["Symbol"]
print(tickers)
```

```python
import yfinance as yf
import pandas as pd

# Define date range
start_date = "2009-12-01"
end_date = "2025-09-30"

# Replace dots with hyphens in tickers (Yahoo Finance format)
tickers_clean = [ticker.replace('.', '-') for ticker in tickers]

# Combine S&P 500 ticker with your tickers
sp_500 = "^GSPC"
all_tickers = [sp_500] + tickers_clean

# Download data with auto_adjust explicitly set
df_data = yf.download(all_tickers, start=start_date, end=end_date, auto_adjust=True)['Close']

print(df_data.head())
print(f"\nShape: {df_data.shape}")
```

```python
# Check number of missing values per column
print(df_data.isnull().sum())

# Or to see only columns that have missing values
print(df_data.isnull().sum()[df_data.isnull().sum() > 0])

# To see percentage of missing data
print((df_data.isnull().sum() / len(df_data) * 100).round(2))

# To see columns that are completely empty
empty_columns = df_data.columns[df_data.isnull().all()].tolist()
print(f"\nCompletely empty columns: {empty_columns}")
```

```python
# Get columns with no missing values
columns_complete = df_data.columns[df_data.isnull().sum() == 0].tolist()

# Get columns with missing values
columns_with_missing = df_data.columns[df_data.isnull().sum() > 0].tolist()

print(f"Complete columns: {len(columns_complete)}")
print(f"Columns with missing data: {len(columns_with_missing)}")
print(f"\nColumns with missing data: {columns_with_missing}")
```

```python
# Drop columns with missing data
df_data = df_data.drop(columns=columns_with_missing)

print(f"New shape: {df_data.shape}")
```

```python
# Reset index to make Date a regular column instead of index
df_data = df_data.reset_index()

# Rename ^GSPC to sp_500
df_data = df_data.rename(columns={'^GSPC': 'sp_500'})
```
